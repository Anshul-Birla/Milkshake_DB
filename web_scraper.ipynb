```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
import time
```

```python
links = []
base = 'https://cookpad.com/us/search/milkshakes?page='
for i in range(12):
    links.append(base + str(i + 1))
links
```

```python
recipes = []
driver = webdriver.Chrome(executable_path = 'C:\Drivers\Chrome_driver\chromedriver')
for link in links:
    driver.get(link)
    time.sleep(3)
    driver.maximize_window()
    data = driver.page_source
    soup = BeautifulSoup(data, 'html.parser')
    temp = soup.find_all('a', class_ = 'media')
    temp = ['https://cookpad.com' + i['href'] for i in temp]
    for i in temp:
        recipes.append(i)
driver.close()
recipes = list(set(recipes))
recipes
```

```python
df = pd.DataFrame(columns = ['Title', 'Ingredients', 'Steps', 'Link'])
driver = webdriver.Chrome(executable_path = 'C:\Drivers\Chrome_driver\chromedriver')
count = 0
driver.get('https://cookpad.com/us/')
time.sleep(120) #for sign in
for recipe in recipes:
    try:
        driver.get(recipe)
        time.sleep(30)
        driver.maximize_window()
        df.at[count, 'Title'] = driver.find_element_by_xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "field-group--no-container-xs", " " ))]').text
        df.at[count, 'Ingredients'] = driver.find_element_by_xpath('//*[(@id = "ingredients")]').text
        df.at[count, 'Steps'] = driver.find_element_by_xpath('//*[(@id = "steps")]').text
        df.at[count, 'Link'] = recipe
        df.to_csv('recipes.csv', index = False)
    except:
        print('Failed at: ' + recipe)
    count += 1
driver.close()
df.head()
```
